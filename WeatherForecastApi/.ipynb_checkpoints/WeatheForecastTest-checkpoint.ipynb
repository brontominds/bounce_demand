```python
def getSeason(dateGive):
    Y = 2000
    if isinstance(dateGive, dt.datetime):
        dateGive = dateGive.date()
    else:
        raise TypeError("wrong Type Given")
    dateGive = dateGive.replace(year=Y)
    return next((season,code) for start, end,season, fancy,code in seasons
                if start <= dateGive <= end)
```

```python
weather_forecast['season'], weather_forecast['season_code'] = zip(*weather_forecast['timestamp_local'].apply(getSeason))
```

```python
def checkHol(dateGive):
    if isinstance(dateGive, dt.datetime):
        dateGive = dateGive.date()
        
    if not isinstance(dateGive, dt.date):
        raise TypeError("wrong Type Given")
        
    if dateGive in holidays:
        return 1
    return 0
```

```python
weather_forecast['holiday'] = weather_forecast['timestamp_local'].dt.date.apply(checkHol)
```
